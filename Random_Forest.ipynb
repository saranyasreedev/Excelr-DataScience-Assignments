# Random Forest

Since the target variable is continious, we create a class of the value based on the mean

<= 7.49 == "Small" and > 7.49 == "large"

```python
data1["sales"]="small"
data1
```

```python
data1.loc[data1["Sales"]>7.49,"sales"]="large"
data1
```

```python
data1.drop(["Sales"],axis=1,inplace=True)
data1
```

```python
X = data1.iloc[:,0:14]
Y = data1.iloc[:,14]
```

```python
X
```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=0)
```

```python
model = RF(n_estimators=10, max_depth = 3) 
model.fit(x_train,y_train)
```

```python
pred_train = model.predict(x_train)
```

```python
pred_train
```

```python
accuracy_score(y_train,pred_train)
```

```python
confusion_matrix(y_train,pred_train)
```

```python
pred_test = model.predict(x_test)
```

```python
pred_test
```

```python
accuracy_score(y_test,pred_test)
```

```python
confusion_matrix(y_test,pred_test)
```

```python
df = pd.DataFrame({'Actual':y_test,'Predicted':pred_test})
df
```

# Conclusion

Since the accuracy of the Training set is 100% we test the accurancy on the test data which is 80%

As seen in the confusion matrix of Test data 32 instances are presdected correctly and 8 instances are not
